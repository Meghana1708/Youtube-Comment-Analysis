```python
class CustomDataset(Dataset):
    def __init__(self, text, labels, tokenizer, max_len):
        self.text = text
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.text)

    def __getitem__(self, idx):
        text = str(self.text[idx])
        label = self.labels[idx]

        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            return_token_type_ids=False,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt',
        )

        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }

# Create instances of the custom dataset class
max_len = 128  # You can adjust this based on your text length distribution
train_dataset = CustomDataset(
    text=train_df['commentText'].values,
    labels=train_df['encoded_sentiment'].values,
    tokenizer=tokenizer,
    max_len=max_len
)

test_dataset = CustomDataset(
    text=test_df['commentText'].values,
    labels=test_df['encoded_sentiment'].values,
    tokenizer=tokenizer,
    max_len=max_len
)

# Define data loaders
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

# Define optimizer and loss function
optimizer = AdamW(model.parameters(), lr=2e-5)
loss_fn = torch.nn.CrossEntropyLoss()

# Train the model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
num_epochs = 2 # You can adjust the number of epochs
for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for batch in tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs}'):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)

        optimizer.zero_grad()

        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        loss.backward()
        optimizer.step()

    average_loss = total_loss / len(train_loader)

    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {average_loss}')
```

```python
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for batch in tqdm(test_loader, desc='Evaluating'):
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits

        preds = torch.argmax(logits, dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Decode labels back to original sentiment categories
decoded_preds = label_encoder.inverse_transform(all_preds)
decoded_labels = label_encoder.inverse_transform(all_labels)

# Report metrics
print(classification_report(decoded_labels, decoded_preds))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import pandas as pd
test_df['Predicted_Label'] = decoded_preds
# Assuming you already have the true and predicted labels in these lists
true_labels = test_df['Labels'].tolist()  # Ground truth labels
predicted_labels = test_df['Predicted_Label'].tolist()  # Model predictions
categories = sorted(list(set(true_labels)))
# 1. Confusion Matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Plot Confusion Matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=categories, yticklabels=categories)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for BERT Model')
plt.show()

# 2. Classification Report Visualization (Precision, Recall, F1)
# Get precision, recall, and f1-score from classification_report
report = classification_report(true_labels, predicted_labels, target_names=categories, output_dict=True)
df_report = pd.DataFrame(report).transpose()

# Bar plot for precision, recall, and F1-score
df_report[['precision', 'recall', 'f1-score']].iloc[:-3].plot(kind='bar', figsize=(10, 6))
plt.title('Precision, Recall, and F1-Score for Each Class')
plt.xlabel('Class Labels')
plt.ylabel('Score')
plt.xticks(rotation=45, ha='right')
plt.show()

# 3. Correctly Predicted Samples for Each Label
correct_predictions = np.diag(cm)  # Diagonal values are correctly predicted samples
total_samples = cm.sum(axis=1)  # Total samples in each class (row sums of confusion matrix)
incorrect_predictions = total_samples - correct_predictions  # Incorrectly predicted samples

# Display number of correctly and incorrectly predicted samples for each label
for i, label in enumerate(categories):
    print(f"Label: {label}")
    print(f"Correct Predictions: {correct_predictions[i]}")
    print(f"Incorrect Predictions: {incorrect_predictions[i]}")
    print(f"Total Samples: {total_samples[i]}")
    print(f"Accuracy for {label}: {correct_predictions[i] / total_samples[i]:.2f}\n\n\n")

# 4. Plot of Correct vs Incorrect Predictions for Each Label
correct_vs_incorrect = pd.DataFrame({
    'Correct Predictions': correct_predictions,
    'Incorrect Predictions': incorrect_predictions
}, index=categories)

# Plot Correct vs Incorrect Predictions
correct_vs_incorrect.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'red'])
plt.title('\n\nCorrect vs Incorrect Predictions for Each Class')
plt.xlabel('Class Labels')
plt.ylabel('Number of Predictions')
plt.xticks(rotation=45, ha='right')
plt.show()
```
